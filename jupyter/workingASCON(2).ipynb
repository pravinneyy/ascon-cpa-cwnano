```python
import numpy as np
from tqdm import tqdm
import os 
import time 

num_traces = 3000 
num_samples = 2048
scope.adc.samples = num_samples

secret_key = [
    0xB7, 0xE3, 0xE9, 0x5C,
    0x22, 0xD4, 0x71, 0xBE,
    0x09, 0xF2, 0x67, 0x3A,
    0x8D, 0xC8, 0x54, 0xEF
]

trace_array = np.zeros((num_traces, num_samples))
nonce_array = np.zeros((num_traces, 16), dtype=np.uint8)

print("Starting trace capture with straight 157-byte zero padding...")
target.flush()

ascon_flags = 0x38

for i in tqdm(range(num_traces), desc="Capturing traces"):
    random_nonce = list(os.urandom(16))
    nonce_array[i] = random_nonce
    padded_payload = [ascon_flags] + random_nonce + secret_key + [0x00] * 157
    scope.arm()
    
    target.simpleserial_write('a', bytearray(padded_payload))
    
    if scope.capture():
        print(f"\n❌ Timeout or trigger error at trace {i}!")
        break
        
    _ = target.simpleserial_read('r', 96)
    
    trace_array[i] = scope.get_last_trace()

print("\n✔ Capture complete!")
print(f"Captured trace shape: {trace_array.shape}")
```

```python
IV_BYTES = np.array(
    [0x80, 0x40, 0x0C, 0x06, 0x00, 0x00, 0x00, 0x00],
    dtype=np.uint8
)

HW_TABLE = np.array(
    [bin(x).count("1") for x in range(256)],
    dtype=np.float32
)

def build_hypothesis_matrix(byte_idx, nonce_array):
    x0 = int(IV_BYTES[byte_idx])

    x3 = nonce_array[:, byte_idx].astype(np.int32)[:, np.newaxis]
    x4 = nonce_array[:, byte_idx + 8].astype(np.int32)[:, np.newaxis]

    kg = np.arange(256, dtype=np.int32)[np.newaxis, :]

    y4_bytes = (
        (x4 & kg)
        ^ x4
        ^ x3
        ^ (kg & x0)
        ^ kg
    ) & 0xFF

    return HW_TABLE[y4_bytes]

def pearson_correlation(H, traces):
    H_zm = H      - H.mean(axis=0, keepdims=True)
    T_zm = traces - traces.mean(axis=0, keepdims=True)
    numerator   = H_zm.T @ T_zm
    H_std       = np.sqrt((H_zm ** 2).sum(axis=0))
    T_std       = np.sqrt((T_zm ** 2).sum(axis=0))
    denominator = np.outer(H_std, T_std)
    denominator[denominator == 0] = 1e-12
    return numerator / denominator 

def cpa_attack(trace_array, nonce_array, known_key=None):
    num_traces, num_samples = trace_array.shape

    print(f"\n{'='*60}")
    print(f"CPA Attack — ASCON Round 1")
    print(f"Traces: {num_traces}   Samples per trace: {num_samples}")
    print(f"{'='*60}\n")

    recovered = []
    all_max_corr = []

    for byte_idx in range(8):

        H = build_hypothesis_matrix(byte_idx, nonce_array)

        corr = pearson_correlation(H, trace_array)

        max_corr = np.max(
            np.abs(corr),
            axis=1
        )

        best_kg = int(
            np.argmax(max_corr)
        )

        recovered.append(best_kg)
        all_max_corr.append(max_corr)

        line = (
            f"Byte {byte_idx}: "
            f"0x{best_kg:02X} "
            f"(corr={max_corr[best_kg]:.4f})"
        )

        if known_key is not None:
            ck = known_key[byte_idx]

            ok = "✔" if best_kg == ck else "✗"

            line += (
                f" {ok} "
                f"(correct=0x{ck:02X}, "
                f"corr={max_corr[ck]:.4f})"
            )

        print(line)

    print("\nRecovered K0:")
    print(" ".join(f"{b:02X}" for b in recovered))

    return recovered, all_max_corr
    
trace_window = trace_array[:, 1400:2050]
peak = 340
trace_attack = trace_window[:, peak-20:peak+20]

recovered, all_max_corr = cpa_attack(
    trace_attack,
    nonce_array,
    known_key=secret_key[0:8],
)


```

```python
peak_y2 = 1660
trace_attack_y2 = trace_array[:, peak_y2-20:peak_y2+20]

def build_hypothesis_matrix_y2(byte_idx, nonce_array, known_x1_byte):
    x1 = known_x1_byte
    x3 = nonce_array[:, byte_idx].astype(np.int32)[:, np.newaxis]
    x4 = nonce_array[:, byte_idx + 8].astype(np.int32)[:, np.newaxis]
    kg = np.arange(256, dtype=np.int32)[np.newaxis, :]

    y2_bytes = (
        (x4 & x3)
        ^ x4
        ^ kg
        ^ x1
        ^ 0xFF
    ) & 0xFF

    return HW_TABLE[y2_bytes]

def cpa_attack_y2(trace_array, nonce_array, recovered_x1, known_key=None):
    num_traces, num_samples = trace_array.shape
    print(f"\n{'='*60}")
    print(f"CPA Attack — ASCON Round 1 — target y2 (key[8:16])")
    print(f"Traces: {num_traces}   Samples per trace: {num_samples}")
    print(f"{'='*60}\n")

    recovered    = []
    all_signed_peak = []

    for byte_idx in range(8):
        H    = build_hypothesis_matrix_y2(byte_idx, nonce_array, recovered_x1[byte_idx])
        corr = pearson_correlation(H, trace_array)  
        peak_sample_per_kg = np.argmax(np.abs(corr), axis=1)           
        signed_peak        = corr[np.arange(256), peak_sample_per_kg]   

     
        best_kg  = int(np.argmax(signed_peak))
        best_val = signed_peak[best_kg]

        sorted_vals = np.sort(signed_peak)[::-1]
        margin = sorted_vals[0] - sorted_vals[1]

        recovered.append(best_kg)
        all_signed_peak.append(signed_peak)

        line = f"Byte {byte_idx}: 0x{best_kg:02X} (signed_corr={best_val:.4f}, margin={margin:.4f})"
        if known_key is not None:
            ck = known_key[8 + byte_idx]
            ok = "✔" if best_kg == ck else "✗"
            line += f"  {ok} (correct=0x{ck:02X}, its signed_corr={signed_peak[ck]:.4f})"
        print(line)

    print("\nRecovered K1 (key[8:16]):")
    print(" ".join(f"{b:02X}" for b in recovered))
    return recovered, all_signed_peak


recovered_x2, all_signed_peak_x2 = cpa_attack_y2(
    trace_attack_y2,
    nonce_array,
    recovered_x1=recovered,
    known_key=secret_key[:],
)
```
